```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
tf = np.linspace(0.1, 20, 100)

# BM25 Query Term Frequency (k_3=1) : ((k_3 + 1) * tf) / (k_3 + tf)
k_3 = 1

plt.plot(tf, ((k_3 + 1) * tf) / (k_3 + tf), label='BM25 TF')
# Square root TF
plt.plot(tf, np.sqrt(tf), label='Sqrt TF')
# Log TF
plt.plot(tf, np.log(tf + 1), label='Log TF')
# # Raw TF
# plt.plot(tf, tf, label='Raw TF')

plt.grid(True)
plt.xlabel('Term Frequency')
plt.ylabel('TF Score')

plt.legend()
plt.show()
```

```python
tf = np.linspace(0.1, 50, 100)

# BM25 Term Frequency (k_3=1) : ((k_3 + 1) * tf) / (k_3 + tf)
k_3 = 1
plt.plot(tf, ((k_3 + 1) * tf) / (k_3 + tf), label='k1=1')
k_3 = 10
plt.plot(tf, ((k_3 + 1) * tf) / (k_3 + tf), label='k1=10')
k_3 = 100
plt.plot(tf, ((k_3 + 1) * tf) / (k_3 + tf), label='k1=100')


plt.grid(True)
plt.xlabel('Term Frequency')
plt.ylabel('TF Score')

plt.title("BM25 TF score curve under different k1")
plt.legend()
plt.show()
```

```python
L = np.linspace(1, 1000, 100)

f_dt = 100
k_1 = 1.5
L_avg = 1000

b = 0
plt.plot(L, ((k_1 + 1) * f_dt) / (k_1 * (1-b + b * L/L_avg) + f_dt), label='b=0')
b = 0.5
plt.plot(L, ((k_1 + 1) * f_dt) / (k_1 * (1-b + b * L/L_avg) + f_dt), label='b=0.5')
b = 1
plt.plot(L, ((k_1 + 1) * f_dt) / (k_1 * (1-b + b * L/L_avg) + f_dt), label='b=1')

plt.grid(True)
plt.xlabel('Document Length')
plt.ylabel('TF Score')

plt.legend()
plt.show()
```

```python
f_t = np.linspace(0, 200, 100)

N = 10000
plt.plot(f_t, np.log((N - f_t + 0.5)/(f_t + 0.5)), label='IDF')
# plt.plot(f_t, np.sqrt((N - f_t + 0.5)/(f_t + 0.5)), label='sqrt-IDF')

plt.grid(True)
plt.xlabel('Document Frequency')
plt.ylabel('TF Score')

plt.title("IDF curve when number of documents fixed at 10000")
plt.legend()
plt.show()
```

```python
import nltk
nltk.download('wordnet')

sentence = ['Ethiopian', 'Airlines', ':', 'Boeing', 'faces', 'questions', 'after', 'crash', '.']
lemmatiser = nltk.stem.wordnet.WordNetLemmatizer()
stemmer = nltk.stem.porter.PorterStemmer()

# Code below from ...
def lemmatise(word):
    lemma = lemmatiser.lemmatize(word, 'v')
    if lemma == word:
        lemma = lemmatiser.lemmatize(word, 'n')
    return lemma
# End of copied code

lemmatised_sent = [lemmatise(word) for word in sentence ]
stemmed_sent = [stemmer.stem(word) for word in sentence ]

print('Sentence after lemmatisation: 'lemmatised_sent)
print('Sentence after stemming: ', stemmed_sent)
```
